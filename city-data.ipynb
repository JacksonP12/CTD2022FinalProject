```python
#import dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#import datasets
#carbon emissions data (modified from million metric tons to metric tons): https://www.eia.gov/environment/emissions/state/
#population data: https://www.census.gov/data/tables/time-series/demo/popest/2010s-total-cities-and-towns.html
tport = pd.read_csv('transportation.csv')
popu = pd.read_csv('citypop.csv')
print(tport.head())
print(popu.head())
```

```python
#filter to Unlinked Passenger Trips & Vehicle Revenue Miles
tport = tport[['City', 'Vehicle Revenue Miles']]
tport
```

```python
#merge state values
tport = tport.groupby('City')[['Vehicle Revenue Miles']].sum()
tport
```

```python
#merge datasets (2018)
full_df = tport.merge(popu, on='City')
full_df = full_df[['City', 'Vehicle Revenue Miles', '2018']]
full_df.rename(columns = {'2018' : 'Population'}, inplace = True)
full_df = full_df.groupby('City')[['Vehicle Revenue Miles', 'Population']].sum().reset_index()
full_df
```

```python
full_df.to_csv('city_data.csv')
```
